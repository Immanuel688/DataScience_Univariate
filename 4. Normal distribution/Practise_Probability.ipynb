```python
# Distribution of range or area under curve, visuals, sampling
from matplotlib import pyplot
from scipy.stats import norm
import seaborn as sns
graph = sns.distplot(dataset['ssc_p'],kde=True,kde_kws={'color':'blue'},color='Green')
pyplot.axvline(60,color='Red')
pyplot.axvline(80,color='Red')
sample=dataset['ssc_p']
sample_mean =sample.mean()
sample_std = sample.std()
print(f'Mean={sample_mean:.3f}, standard deviation ={sample_std:.3f}')
dist = norm(sample_mean, sample_std)
values=[]
for value in range(60, 80):
    values.append(value)
probabilities=[]
for value in range(60, 80):
    probabilities.append(dist.pdf(value))
prob= sum(probabilities)
print(f'The area between range (60,80) is : {prob}')
```

```python
# CREATING FUNCTION
def pdf_prob(dataset,columnName, startrange, endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    graph = sns.distplot(dataset[columnName],kde=True,kde_kws={'color':'blue'},color='Green')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    sample=dataset[columnName]
    sample_mean =sample.mean()
    sample_std = sample.std()
    print(f'Mean={sample_mean:.3f}, standard deviation ={sample_std:.3f}')
    values=[]
    for value in range(startrange, endrange):
        values.append(value)
    probabilities=[]
    for value in range(startrange, endrange):
        probabilities.append(dist.pdf(value))
    prob= sum(probabilities)
    print(f'The area between range ({startrange},{endrange}) is : {prob}')
    return prob
```
