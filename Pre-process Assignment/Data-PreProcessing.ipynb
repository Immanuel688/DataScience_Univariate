```python
#Acc to above error we have to use this only for quantitative dataset
def quanqual(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtypes=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual= quanqual(dataset)
```

```python
quan
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(missing_values= np.nan, strategy="constant", fill_value=0)
new_dataset=imp.fit(dataset[quan])
new_dataset=imp.transform(dataset[quan])
```

```python
new_dataset
```

```python
new_dataset=pd.DataFrame(new_dataset,columns=quan)
new_dataset
```

```python
dataset[qual]
```

```python
combined_data = pd.concat([new_dataset, dataset[qual]],axis=1)
combined_data
```

```python
combined_data.isnull().sum()
```

```python
combined_data.to_csv('Preprocessed_PlacementInput.csv', index= False)
```
