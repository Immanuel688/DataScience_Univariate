```python
import pandas as pd
dataset=pd.read_csv("kidney_disease_train.csv")
dataset
```

```python
dataset.isnull()
```

```python
dataset.isnull().sum()
```

```python
def quanqual(dataset):
    qual=[]
    quan=[]
    for columnName in dataset.columns:
        if(dataset[columnName].dtypes=='O'):
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan,qual
quan,qual= quanqual(dataset)
```

```python
quan
```

```python
qual
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp= SimpleImputer(missing_values= np.nan, strategy='mean')
new_dataset1=imp.fit(dataset[quan])
new_dataset1=imp.transform(dataset[quan])
```

```python
new_dataset1=pd.DataFrame(new_dataset1,columns=quan)
new_dataset1
```

```python
imp2= SimpleImputer(missing_values= np.nan, strategy='most_frequent')
new_dataset2=imp2.fit(dataset[qual])
new_dataset2=imp2.transform(dataset[qual])
```

```python
new_dataset2=pd.DataFrame(new_dataset2,columns=qual)
new_dataset2
```

```python
# Combine imputed numerical and categorical dataframes
combined_data = pd.concat([new_dataset1, new_dataset2],axis=1)
combined_data
```

```python
combined_data.isnull().sum()
```

```python
def splitting(dataset,quan,qual):
    import numpy as np
    from sklearn.impute import SimpleImputer
    imp= SimpleImputer(missing_values= np.nan, strategy='mean')
    new_dataset1=imp.fit_transform(dataset[quan])
    new_dataset1=pd.DataFrame(new_dataset1,columns=quan)
    
    imp= SimpleImputer(missing_values= np.nan, strategy='most_frequent')
    new_dataset2=imp.fit_transform(dataset[qual])
    new_dataset2=pd.DataFrame(new_dataset2,columns=qual)
    
    combined_data = pd.concat([new_dataset1, new_dataset2],axis=1)
    combined_data
    return combined_data
    
```

```python
splitting(dataset,quan,qual)
```
